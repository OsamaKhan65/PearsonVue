# Advantages of Python:

1. **Ease of Learning and Readability:** Python has a clean and straightforward syntax that makes it easy to learn, read, and write. It emphasizes readability, reducing the cost of program maintenance and development.
2. **Large Standard Library:** Python comes with a vast standard library that includes modules for a wide range of tasks, making it convenient for developers to perform various tasks without having to build everything from scratch.
3. **Community and Third-party Libraries:** Python has a large and active community that contributes to an extensive ecosystem of third-party libraries and frameworks. These libraries cover diverse domains, such as web development, data analysis, machine learning, and more, enabling developers to leverage existing solutions easily.
4. **Cross-platform Compatibility:** Python is a platform-independent language, meaning code written on one platform can run on other platforms without modification, making it highly versatile for different operating systems.
5. **Versatility:** Python can be used for a wide variety of applications, from web development to scientific computing, data analysis, artificial intelligence, automation, and more.
6. **Rapid Prototyping and Development:** The language's simplicity and high-level nature facilitate rapid prototyping and development, allowing developers to iterate quickly on their projects.
7. **Dynamically Typed:** Python is dynamically typed, which means you don't need to declare variable types explicitly. This can speed up development but may require extra care to catch type-related errors during runtime.

# Disadvantages of Python:

1. **Speed:** Python is an interpreted language, and as such, it can be slower compared to compiled languages like C or C++. While this performance gap has been reduced with advancements like PyPy and Just-In-Time (JIT) compilers, Python might not be the best choice for highly performance-critical applications.
2. **Global Interpreter Lock (GIL):** Python's GIL can limit multi-core processor performance for certain CPU-bound tasks, as only one thread can execute Python bytecode at a time. However, this issue can be mitigated by using multi-processing or threading libraries that work around the GIL.
3. **Mobile Development:** Python is not the first choice for mobile app development, as languages like Java (for Android) and Swift/Objective-C (for iOS) are more commonly used for native mobile app development.
4. **Static Typing Limitations:** Although Python has introduced optional type hints in recent versions, it is still dynamically typed by default. This can lead to certain type-related bugs that would have been caught at compile-time in statically typed languages.
5. **Memory Consumption:** Python can consume more memory compared to some other languages due to its high-level abstractions and dynamic nature. This may not be ideal for certain memory-constrained environments.
6. **Less Suitable for Low-Level Systems Programming:** While Python is versatile, it may not be the best choice for low-level systems programming where direct hardware access or memory management is crucial. Low-level languages like C or Rust are more appropriate for such tasks.

3. Python relies on indentation, using whitespace, to define scope; such as the scope of loops, functions and classes. Other programming languages often use curly-brackets for this purpose.

# Running Python Scripts:

Now that you have setup python on your computer, we can start programming. There are three ways to execute python programs:

#### Using Command Line Interface (CLI):
The first method is to directly write code in command prompt or terminal. In this way, the code gets executed line by line, but you cannot save it for later use. To write a program this way, follow these steps:

1. Open Command Prompt.
2. Write "python" and hit enter. (without double quotes)
3. Write "print('Hello World!')" and hit enter. (without double quotes)

You will see "Hello World!" written in the next line.



#### Using Text Editor:

In order to save a script for later use, you can follow these steps:

1. Open Notepad or any other text editor.
2. Write "print('Hello World!')". (without double quotes)
3. Save the file as python script using ".py" extension. (you can close the file)
4. Open Command Prompt.
5. Write "python /path/to/py/file/name.py" and hit enter. (without double quotes)

You will see "Hello World!" written in the next line.



#### Using Integrated Development Environment (IDE):

The most widely used method of executing python scripts is using IDEs. IDEs are a type of software application that provides you with various advanced features in addition to letting you write and run scripts, like debugging, library handeling, version handeling, virtual environments, integration with other softwares and much more.

You can choose your favorite IDE for programming in python based on your needs and preferences. There are python IDEs ranging from basic (IDLE) to mid-range (Spyder) to some really advanced ones (PyCharm, VSCode). There is also another type of IDE that lets you create notebooks with executable cells instead of writing a script like this one (Jupyter Notebook, Google Colab).
